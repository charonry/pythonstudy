```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib import rcParams
```

```python
rcParams['font.family'] = 'SimHei'
rcParams['axes.unicode_minus'] = False  # 关闭unicode负号，改用ASCII负号
```

```python
plt.figure(figsize=(9,3))
month = ['1月','2月','3月','4月']
sales = [100,140,80,120]
plt.plot(month,sales,label='产品A',color='orange',linewidth=2,marker='o',linestyle='--')
# 标签
plt.title("标题",color='red',fontsize=20)
# 标签
plt.xlabel('月份',fontsize=10)
plt.ylabel('销售额',fontsize=10)
# 图例
plt.legend(loc="upper left")
# 添加网格线
plt.grid(axis='y',alpha=0.3,color='grey',linestyle='--')
# 设置刻度相关
plt.xticks(rotation=45,fontsize=10)
plt.yticks(rotation=0,fontsize=10)
# y轴范围设置
plt.ylim(0,150)
# 每个数据点显示数值
for x,y in zip(month,sales):
    plt.text(x,y,str(y),ha='center',va='bottom',fontsize=10)
plt.show()
```

```python
plt.figure(figsize=(9,3))
subjects = ['语文', '数学', '英语']
scores = [85, 92, 88]
plt.bar(subjects, scores,label='张三',width=0.5)
plt.title("标题",color='red',fontsize=20)
plt.xlabel('学科',fontsize=10)
plt.ylabel('分数',fontsize=10)
plt.legend(loc="upper right")
plt.grid(axis='y',alpha=0.3,color='grey',linestyle='--')
plt.xticks(rotation=45,fontsize=10)
plt.yticks(rotation=0,fontsize=10)
plt.ylim(0,100)
for x,y in zip(subjects,scores):
    plt.text(x,y,str(y),ha='center',va='bottom',fontsize=10)
# 自动优化排版
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(9,3))
subjects = ['语文', '数学', '英语']
scores = [85, 92, 88]
plt.barh(subjects, scores,label='张三',)
plt.title("标题",color='red',fontsize=20)
plt.xlabel('分数',fontsize=10)
plt.ylabel('学科',fontsize=10)
plt.xlim(0,100)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(9,3))
things = ['游泳','网球','跑步',]
times = [2,3,1]
colors = ['#66b3ff','#99ff99','#ffcc99']
explode = [0,0,0.1]
plt.pie(times,labels=things,
        autopct='%1.1f%%',  # autopct:展示数据格式
        startangle=60,  # startangle:调整角度
        colors=colors,
        pctdistance=0.6,  # pctdistance:数字远离圆心占比
        wedgeprops={'width':0.6}, # wedgeprops:圆环空心占比;
        explode=explode,  # 设置突出块的位置
        shadow=True,  # 阴影
        )
plt.title("标题",color='red',fontsize=20)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(9,5))
x = np.random.uniform(0, 10,800)
y = 2*x + np.random.normal(0,2,800)
plt.scatter(x,y,
            color='blue',
            alpha=0.5,  # 透明度
            s=10,  # 圆点大小
            label='数据'
            )
plt.plot([0,10],[0,20],color='orange',linestyle='--',linewidth=2,label='基准')
plt.title("标题",color='red',fontsize=15)
plt.xlabel('X变量',fontsize=10)
plt.ylabel('Y变量',fontsize=10)
plt.legend(loc="upper left")
plt.grid(axis='y',alpha=0.3,color='grey',linestyle='--')
plt.ylim(0,30)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.tight_layout()
plt.show()
```

```python
data = {
    '语文':[82,85,88,70,90,76,84,83,95],
    '数学':[75,80,79,70,93,88,82,87,92],
    '英语':[70,72,68,65,78,80,85,90,95]
}
plt.figure(figsize=(8,3))
# 箱线图
# Q3 75%值; Q2 50%值; Q1 25%值
# IQR = Q3-Q1 :四分位间距
# 上边缘线：Q3-1.5*IQR
# 下边缘线：Q1-1.5*IQR
plt.boxplot(data.values(),tick_labels=data.keys())
plt.title("标题（箱线图）")
plt.xlabel('学科')
plt.ylabel('成绩')
plt.grid(axis='y',alpha=0.3,color='grey',linestyle='--')
plt.tight_layout()
plt.show()
```
